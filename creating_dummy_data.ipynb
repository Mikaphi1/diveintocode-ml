#**[Problem 1]**
##***Creating random numbers***


```python
import numpy as np

mean = (-3,0)
cov_matrix = np.array([[1.0,0.8],[0.8,1.0]])

np.random.seed(0)
random_numbers = np.random.multivariate_normal(mean,cov_matrix,size=500)

print(random_numbers.ndim)
print(random_numbers.shape)
print(random_numbers)
```

#**[Problem 2]**
##***Visualization by scatter plot***

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("Scatter plot")
plt.xlabel("1x")
plt.ylabel("2x")
plt.scatter(random_numbers[:,0],random_numbers[:,1])
plt.show()
```

#**[Problem 3)**
##***Visualization by histogram***

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("Histogram of 1x")
plt.xlabel("1x")
plt.ylabel("2x")
plt.xlim(-5,2)
plt.hist(random_numbers[:,0], bins=50)
plt.show()

plt.title("Histogram of 2x")
plt.xlabel("1x")
plt.ylabel("frequency")
plt.xlim(-5,2)
plt.hist(random_numbers[:,1], bins=50)
plt.show()
```

#**[Problem 4]**
##***Addition of data***

```python
import numpy as np

mean_1 = (0,-3)
cov_matrix2 = np.array([[1.0,0.8],[0.8,1.0]])

np.random.seed(0)
random_numbers2 = np.random.multivariate_normal(mean_1,cov_matrix2,size=500)

print(random_numbers2.ndim)
print(random_numbers2.shape)
print(random_numbers2)
```

```python
# Visualize the data
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("Scatter plot")
plt.xlabel("1x")
plt.ylabel("2x")
plt.scatter(random_numbers[:,0],random_numbers[:,1], label="0")
plt.scatter(random_numbers2[:,0],random_numbers2[:,1], label="1")
plt.legend()
plt.show()
```

#**[Problem 5]**
#***Data combination***

```python
# Using np.concatenate to combine data
random_numbers_concatenated = np.concatenate((random_numbers, random_numbers2), axis=0)
print(random_numbers_concatenated)
print(random_numbers_concatenated.shape)
```

```python
# Using np.vstack() to combine data
random_numbers_stacked = np.vstack((random_numbers,random_numbers2))
print(random_numbers_stacked)
print(random_numbers_stacked.shape)
```

##***Labeling***

```python
# Create array of 500 zeros
label_0 = np.zeros((500,1))

# Create array of 500 ones
label_1 = np.ones((500,1))

# append ones to zeros
labels = np.concatenate((label_0,label_1),axis=0)

# Add labels to data
random_numbers_labelled = np.concatenate((random_numbers_concatenated,labels), axis=1)

# Check data 
print(random_numbers_labelled)
print(random_numbers_labelled.shape)
```
