#**[Problem 1]**
##***Linear Function***

```python
import numpy as np

x_ndarray = np.arange(-50,50.1,0.1).reshape(1001,1)
y_ndarray = 0.5*x_ndarray + 1
print(x_ndarray)
print(y_ndarray)
```

#**[Problem 2]**
##***Array combionation***

```python
# Combine x_ndarray and y_ndarray
xy_ndarray = np.concatenate((x_ndarray,y_ndarray), axis=1)
print(xy_ndarray)
```

#**[Problem 3]**
###***Find the gradient***

```python
x_0 = xy_ndarray[:-1,0].reshape(1000,1)
x_1 = xy_ndarray[1:,0].reshape(1000,1)

x_change = x_1 - x_0

y_0 = xy_ndarray[:-1,1].reshape(1000,1)
y_1 = xy_ndarray[1:,1].reshape(1000,1)

y_change = y_1 - y_0

gradient = y_change/x_change

print(gradient)
```

#**[Problem 4]**
##***Draw a graph***

```python
import matplotlib.pyplot as plt
%matplotlib inline

# For x against y
plt.title("Linear Function y=1/2x + 1")
plt.xlabel("x")
plt.ylabel("y")

# A plot of x vs y
plt.plot(xy_ndarray[:,0],xy_ndarray[:,1], label ="x vs y",color="blue",linewidth=3.5)

# A plot of x vs the gradient
plt.plot(xy_ndarray[1:,0],gradient, label="x vs gradient",color="red",linewidth=3.5)

plt.legend()
plt.show()
```

#**[Problem 5**
##***Python functionalization***

```python
import numpy as np

# For y = x^2
def function1(x):
  return x**2

#create gradient function
def gradient_compute(function, x_range):
  import numpy as np
  x = x_range
  y = function(x_range)
  xy_array = np.vstack([x,y]).T
  x_difference = x[1:] - x[:-1]
  y_difference = y[1:] - y[:-1]
  gradient = y_difference/x_difference
  return xy_array, gradient

xy_array1, gradient1 = gradient_compute(function1, np.arange(-50, 50.1, 0.1))
print(xy_array1)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("Function 1")
plt.plot(xy_array1[:,0],xy_array1[:,1], label = "x vs y", color="blue",linewidth=4)
plt.plot(xy_array1[:-1,0],gradient1, label = "x vs gradient", color="crimson",linewidth=4)
plt.legend()
plt.show()

print(xy_array1)
```

```python
import numpy as np

# For y = 2x^2 + 2^x
def function2(x):
  return (2 * x**2) + (2**x)

xy_array2, gradient2 = gradient_compute(function2, np.arange(-50, 50.1, 0.1))

print(xy_array2)

import matplotlib.pyplot as plt
%matplotlib inline

plt.title("Function 2")
plt.plot(xy_array2[:,0],xy_array2[:,1], label = "x vs y", color="blue",linewidth=4)
plt.plot(xy_array2[:-1,0],gradient2, label = "x vs gradient", color="crimson",linewidth=4)
plt.legend()
plt.show()
```

```python
import numpy as np

# For y = sin(x^0.5)
def function3(x):
  import numpy as np
  return np.sin(x**0.5)

xy_array3, gradient3 = gradient_compute(function3, np.arange(0, 50.1, 0.1))

print(xy_array3)

import matplotlib.pyplot as plt
%matplotlib inline

plt.title("Function 2")
plt.plot(xy_array3[:,0],xy_array3[:,1], label = "x vs y", color="blue",linewidth=4)
plt.plot(xy_array3[:-1,0],gradient3, label = "x vs gradient", color="crimson",linewidth=4)
plt.legend()
plt.show()

print(xy_array3.shape)
```

##***Find the minimum value***

```python
import numpy as np

# For y = x^2
yvalues1 = xy_array1[:,1] #create array of y-values
minimum1 = np.amin(yvalues1)
minindex1 = np.argmin(yvalues1)

print("The minimum value of the first function is {}.".format(minimum1))
print("The index of the minimum y-value for the first function is {}.".format((minindex1)))
```
