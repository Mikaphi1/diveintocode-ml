#**[Problem 1]**
##*Number of grains of rice on the 100th day*

```python
def grain_calculator(exp):
  ## Create exponantial function using for statement
  def power(base, exp): 
    res = 1
    for i in range(exp):
       res *= base
    return res
  ## Calculate number of grains received on each day for 100 days
  rice_grains = power(2, exp)
  return rice_grains

def grain_totals(n):
  grain_output = [] # Create an empty list
  ## Use for statement to apend 100 daily values from the grain calculator
  for i in range(0, n):
    grain_output.append(grain_calculator(i))
  return grain_output

# Save values in list
daily_gain = grain_totals(100)

print(daily_gain) # to check the daily gain in grains 
print(len(daily_gain)) # to check that the length of the list is 100

# Determine the accumulated value on the 100th day
accum = 0
for w in daily_gain:
  accum += w
print("*The accumulated number of grains of rice is {}*".format(accum)) # to show the accumulated value on the 100th day
```

```python
# Create list with daily rice totals from day 1 to day 100
daily_totals = [] # create empty list
daily_addition = 0 
for index in daily_gain:
  daily_addition += index
  daily_totals.append(daily_addition)
print(daily_totals)

## Create list of days from day 1 to day 100 which shall be used in the horizontal axis
days = list(range(1,100+1))
print(days) #to confirm list is 1 to 100
print(len(days)) #to confirm there are 100 elements like other lists
```

##Plot line graph

```python
import matplotlib.pyplot as plt
%matplotlib inline
# Display the graph. Title and axis label name.
plt.title("Sorori Shinzaemon Rice Accumulation")
plt.xlabel("number of days")
plt.ylabel("Number of rice grains")
plt.plot(days, daily_gain, label = "Rice grain gains each day")  # Enter the variable name of the list in "List name"
plt.plot(days, daily_totals, label = "Commulative rice grains per day")
plt.legend()
plt.show()
```

```python
#A function that calculates the number of rice grains per day in the rice anecdote of Shinzaemon Sorori
def compute_sorori_shinzaemon(n_days):
  def day_grain_calculator(n_days):
  ## Create exponantial function using for statement
    def pw(base, n_days): 
      r = 1
      for i in range(n_days):
        r *= base
      return r
  ## Calculate number of grains received on each day for 100 days
    day_rice_grains = pw(2, n_days)
    return day_rice_grains
  
  def day_grain_additions(n):
    day_grain_output = []
    ## Use for statement to apend 100 daily values from the grain calculator
    for i in range(0, n):
      day_grain_output.append(day_grain_calculator(i))
    return day_grain_output

  list_n_grains = day_grain_additions(n_days) # List of the number of rice you get one day (list length equals n_days)

  # Create list with daily rice totals from day 1 to day 100
  def accumulations(n_days):
    daily_accumulations = []
    x = 0
    for j in list_n_grains:
      x += j
      daily_accumulations.append(x)
    return daily_accumulations
  
  list_total_grains = accumulations(n_days) # List of total number of rice received by a certain day (list length is equal to n_days)

  print("list_n_grains = ", list_n_grains)
  print("list_total_grains = ", list_total_grains)
  #print("Sorori Shinzaemon Compute: {:,}".format(sum(list_n_grains))) # gives the correct output
  print("Sorori Shinzaemon Compute: {:,} rice gains".format(max(list_total_grains))) #also gives the correct output

  ## Create list of days from day 1 to day n_days which shall be used in the horizontal axis
  days1 = list(range(1, n_days+1))

  import matplotlib.pyplot as plt
  %matplotlib inline

  # Display the graph. Title and axis label name.
  plt.title("Sorori Shinzaemon Rice Grain Compute")
  plt.xlabel("Number of days")
  plt.ylabel("Number of rice grains")
  plt.plot(days1, list_n_grains, label = "Rice grain gains each day")  # Enter the variable name of the list in "List name"
  plt.plot(days1, list_total_grains, label = "Commulative rice grains per day")
  plt.legend()
  plt.show()

# Dive into code is 4 months long, and that translated to 120 days. Therefore, n_days=120
compute_sorori_shinzaemon(n_days=120)
```

##*How many people can live for how many days with the rice we get?*

A single person requires approximately *400 grams* of rice per day. This translates to two meals which include other ingredients to ensure a minimum of *2,100 kilocalories per day*. There are approximately 48 grains of rice in a gram, and this translates to about *19,200 grains* of rice to feed a person per day.

***Source***: https://www.answers.com/Q/How_many_grains_of_rice_to_feed_a_person_for_a_day

```python
#Create a function to calculate how many people and for how many days they can feed
def n_days_to_consume_rice(number_of_people, number_of_rice_grains):
  amt_of_rice_per_day = number_of_people * 19200
  total_days = number_of_rice_grains/amt_of_rice_per_day
  return int(total_days)

# There are four people in the class, therefore:
print("The class will take {:,} days to consume the rice.".format(n_days_to_consume_rice(4, 1329227995784915872903807060280344575)))
```
